```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
# Access the data according to requirement

# df = pd.read_csv("../0.Datasets/processed/processed_car_data.csv")
df = pd.read_csv("../0.Datasets/processed/processed_car_data.csv").iloc[:100,]
# df = pd.read_csv("../0.Datasets/processed/sliced_car_data.csv")
df.head()
```

```python
features = ['odometer','mmr','sellingprice','condition']
x = df[features]
```

```python
x.head()
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled.shape
```

```python
pca = PCA(n_components= 2)
pca_scaled = pca.fit_transform(x_scaled)
```

```python
pca_scaled.shape
```

```python
contri_dummy = pca.components_.T
print(contri_dummy)
```

```python
contribution_while_dimension_reduction = pd.DataFrame(data=contri_dummy, index=x.columns, columns=['pca 1', 'pca 2'])
contribution_while_dimension_reduction
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(contribution_while_dimension_reduction, annot=True, cmap='coolwarm')
plt.title("Heatmap for contribution of data of x in PCA")
plt.show()
```

```python
plt.figure(figsize=(14, 8))
dendrogram(linkage(pca_scaled, method='ward'))

plt.yticks(range(1,18))

plt.title("dendrogram")
plt.xlabel("data points")
plt.ylabel("Euclidean Distance")

plt.grid()
plt.savefig('./Graph/1.Dendrogram.png')
plt.show()
```

```python
# n cluster = 4
```

```python
cluster = AgglomerativeClustering(n_clusters=4, linkage='ward')
```

```python
labels = cluster.fit_predict(pca_scaled)
```

```python
labels
```

```python
plt.figure(figsize=(14,8))
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=labels, cmap='rainbow')

plt.xlabel('pca 1')
plt.ylabel('pca 2')
plt.title('Hierarchical Clustering using pca ')


plt.savefig('./Graph/1.Hierarchical Clustering using pca.png')
plt.show()
```

```python
silhouette_score_ls = []

for k in range(2, 21):
    cluster = AgglomerativeClustering(n_clusters= k, linkage='ward')
    cluster.fit(pca_scaled)
    score = silhouette_score(pca_scaled, cluster.labels_)
    silhouette_score_ls.append(score)
```

```python
silhouette_score_ls
```

```python
plt.figure(figsize=(14,6))

plt.plot(range(2, 21), silhouette_score_ls,)

plt.xlabel('Cluster value')
plt.ylabel('silhouette_score')
plt.title("silhouette_score with cluster size")

plt.xticks(range(2, 21))
plt.grid()

plt.savefig('./Graph/1.silhouette_score_graph.png')
plt.show()
```
